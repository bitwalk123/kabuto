```python
import matplotlib.font_manager as fm
from matplotlib import dates as mdates
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
date_str = "20250815"
code = "7011"
#filename = f"result_{date_str}_{code}.csv"
filename = "result.csv"
df = pd.read_csv(filename, index_col=0)
df.index = pd.to_datetime(df.index)
df
```

```python
FONT_PATH = "../fonts/RictyDiminished-Regular.ttf"
fm.fontManager.addfont(FONT_PATH)

# FontPropertiesオブジェクト生成（名前の取得のため）
font_prop = fm.FontProperties(fname=FONT_PATH)
font_prop.get_name()

plt.rcParams['font.family'] = font_prop.get_name()
plt.rcParams['font.size'] = 16

fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(df["Profit"])
ax.xaxis.set_major_formatter(mdates.DateFormatter("%m/%d"))

fig.suptitle("ChatGPT5 提案の強化学習サンプルによるデイトレ・シミュレーション", fontsize=16)
ax.set_title("三菱重工業 (7011) のティックデータより、建玉 100 株で繰返売買       ", fontsize=14)
ax.grid()
ax.set_ylabel("収益（円）")

plt.tight_layout(pad=0.4, rect=[0, 0, 1.0, 1.0])
plt.savefig(f"report/{date_str}_{code}_torch.png" )
plt.show()
```
