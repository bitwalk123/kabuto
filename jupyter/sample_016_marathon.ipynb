```python
import datetime
import os
import re

import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd
from matplotlib import dates as mdates
```

```python
dt = datetime.datetime.now()
date_str = f"{dt.year:04d}{dt.month:02d}{dt.day:02d}"
date_str
```

```python
code = "7011"
code
```

```python
path_png = os.path.join("..", "screenshots", f"{date_str}_{code}_marathon.png")
path_png
```

```python
file = "result_20251128180846.csv"
path_file = os.path.join("..", "logs", file)
path_file
```

```python
df = pd.read_csv(path_file)
df.head()
```

```python
pattern = re.compile(r"ticks_(\d{4})(\d{2})(\d{2})\.xlsx")

list_index = list()
for file in df["date_str"]:
    if m := pattern.match(file):
        date_str = f"{m.group(1)}-{m.group(2)}-{m.group(3)}"
    else:
        date_str = "1970-01-01"
    list_index.append(date_str)
df.index = pd.to_datetime(list_index)
df.index.name = "Datetime"
df.head()
```

```python
FONT_PATH = "../fonts/RictyDiminished-Regular.ttf"
fm.fontManager.addfont(FONT_PATH)

# FontProperties
font_prop = fm.FontProperties(fname=FONT_PATH)
font_prop.get_name()

plt.rcParams["font.family"] = font_prop.get_name()
plt.rcParams["font.size"] = 9

fig = plt.figure(figsize=(6, 5))
n = 2
ax = dict()
gs = fig.add_gridspec(n, 1, wspace=0.0, hspace=0.0, height_ratios=[1 for i in range(n)])
for i, axis in enumerate(gs.subplots(sharex="col")):
    ax[i] = axis
    ax[i].grid()

ax[0].plot(
    df["total"],
    color="C0",
    linewidth=1,
    label=f" : {df["total"].sum():,.0f}  / {len(df)} ",
)
ax[0].set_ylabel("1 ")
ax[0].xaxis.set_major_formatter(mdates.DateFormatter("%m/%d"))
ax[0].legend(fontsize=7)
ax[0].set_title(f" ({code})")

ax[1].plot(df["trade"], color="C1", linewidth=1)
ax[1].set_ylabel("")
ax[1].xaxis.set_major_formatter(mdates.DateFormatter("%m/%d"))

plt.tight_layout()
plt.savefig(path_png)
plt.show()
```
