```python
import sys

path_up = ".."
if not path_up in sys.path:
    sys.path.append(path_up)
```

```python
import yfinance as yf

from funcs.plot import trend_diff
```

```python
list_code = [
    "AONNY",
    "ITOCY",
    "JXAMY",
    "JXHLY",
    "KXIAY",
    "MHVIY",
    "MUFG",
    "NMR",
    "SFTBY",
    "SONY",
    "TM",
]
for code in list_code:
    symbol = f"{code}"
    yticker = yf.Ticker(symbol)
    df = yticker.history(period="3mo", interval="1d")

    if len(df) > 0:
        df["Diff"] = df["High"] - df["Low"]

    if "shortName" in yticker.info:
        name = yticker.info['shortName']
    elif "longName" in yticker.info:
        name = yticker.info['longName']
    else:
        name = symbol

    trend_diff(code, df, name)
```
