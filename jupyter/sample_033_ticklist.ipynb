```python
import os
import re
import pandas as pd
```

```python
pattern = re.compile(r".+_(\d{4})(\d{2})(\d{2})\..+")

dt_start = pd.to_datetime("2025-08-18")
dt_end = pd.to_datetime("2025-10-01")
path_dir = os.path.join("..", "collection")
list_file_all = sorted(os.listdir(path_dir))
list_file = list()
for file in list_file_all:
    if m := pattern.match(file):
        dt = pd.to_datetime(f"{m.group(1)}-{m.group(2)}-{m.group(3)}")
        if dt_start < dt and dt < dt_end:
            list_file.append(file)
print(list_file)
print(f"\n# of tick files : {len(list_file)}")
```
