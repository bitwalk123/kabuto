# 指定期間で指定銘柄が存在しているティック・ファイル一覧

```python
import sys

path_up = ".."
if not path_up in sys.path:
    sys.path.append(path_up)
```

```python
import datetime
import os
import re

import pandas as pd

from funcs.ios import is_sheet_exists
```

```python
pattern = re.compile(r".+_(\d{4})(\d{2})(\d{2})\..+")

dt_start = pd.to_datetime("2025-08-01")
# dt_end = pd.to_datetime("2025-10-01")
today_str = str(datetime.datetime.now().date())
dt_end = pd.to_datetime(today_str)

code = "9432"  # 銘柄コード
path_dir = os.path.join("..", "collection")
list_file_all = sorted(os.listdir(path_dir))
list_file = list()
for file in list_file_all:
    # 銘柄コードが対象 Excel ファイルのシートに存在するか？
    path_excel = os.path.join(path_dir, file)
    if is_sheet_exists(path_excel, code):
        if m := pattern.match(file):
            dt = pd.to_datetime(f"{m.group(1)}-{m.group(2)}-{m.group(3)}")
            if dt_start < dt and dt < dt_end:
                list_file.append(file)
print(list_file)
print(f"\n# of tick files : {len(list_file)}")
```
